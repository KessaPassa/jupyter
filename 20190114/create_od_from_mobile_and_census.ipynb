```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
root_dir = '/Users/kessapassa/OneDrive/research_log/20190114/'
```

```python
dir_list = ['people10000', 'people20000', 'people30000']
seed_list = [str(123 + i) for i in range(3)]
csv_list = ['mobile']
```

```python
csv_array = {}
for _dir in dir_list:
    csv_array[_dir] = {}
    for _seed in seed_list:
        csv_array[_dir][_seed] = {}
        for _csv in csv_list:
            csv_array[_dir][_seed][_csv] = {}
            df = pd.read_csv(root_dir + 'Origin/' + _dir + 'seed' + _seed + '_' + _csv + '.csv',
                                                    encoding='Shift_JISx0213')
            df = df.loc[:, ['id', 'time', 'area']]
            
            group_list = df.groupby(['id'], sort=True)
            for _id, _group in group_list:
                _group = _group.sort_values(['time'])
                csv_array[_dir][_seed][_csv][_id] = []
                for area in np.asanyarray(_group.loc[:, 'area']):
                    csv_array[_dir][_seed][_csv][_id].append(area)
```

```python
od_array = {}
for _dir in dir_list:
    od_array[_dir] = {}
    for _seed in seed_list:
        od_array[_dir][_seed] = {}
        for _csv in csv_list:
            df = pd.DataFrame()
            for _id in csv_array[_dir][_seed][_csv]:
                csv_array[_dir][_seed][_csv][_id].insert(0, _id)
                tmp = pd.DataFrame(csv_array[_dir][_seed][_csv][_id]).T
                df = pd.concat([df, tmp])
            od_array[_dir][_seed][_csv] = df
            df.to_csv(root_dir + 'od_from_' + _csv + '/' + _dir + 'seed' + _seed + '_' + _csv + '.csv', index=False)
```
