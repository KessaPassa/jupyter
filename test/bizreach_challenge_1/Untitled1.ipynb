```python
argv_dir = {}


def fizzbuzz(n):
    if n > 1:
        fizzbuzz(n - 1)

    result = ''
    for item in argv_dir.items():
        if n % item[0] == 0:
            result += item[1]
    if result == '':
        result = n

    print(result)


def main(argv):
    argv_list = []
    for arg in argv:
        argv_list.append(arg)

    num_range = int(argv_list.pop())

    for arg in argv_list:
        tmp = arg.split(':')
        argv_dir[int(tmp[0])] = tmp[1]

    fizzbuzz(num_range)


main(argv)
```
