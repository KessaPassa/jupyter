```python
import pandas as pd
import numpy as np
import math
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression as LR
```

```python
root_dir = '/Users/kessapassa/OneDrive/research_log/2018_Graduate/2D/'
```

```python
dir_list = ['people10000', 'people20000', 'people30000']
seed_list = [str(123 + i) for i in range(3)]
csv_list = ['census', 'mobile']
```

```python
# エリア番号を線形的な数から、iとjで回した数のようにする
def convert_area_to_contour(area_id):
    area_id = int(area_id)
    contour_id = str(area_id // 6)
    contour_id += str(area_id % 6) + '0'

    return contour_id
```

```python
csv_array = {}
for _dir in dir_list:
    csv_array[_dir] = {}
    for _seed in seed_list:
        csv_array[_dir][_seed] = {}
        for _csv in csv_list:
            df = pd.read_csv(root_dir + _dir + 'seed' + _seed + '_' + _csv + '.csv',
                                                index_col=0,
                                                encoding='Shift_JISx0213')
            df['area'] = df['area'].apply(convert_area_to_contour)
            csv_array[_dir][_seed][_csv] = df
```

```python
csv_array['people10000']['123']['mobile']
```

```python
def get_model(trainX, y):
    model = LR()
    model.fit(trainX, y)
    
    return model
```

```python
model = None

pred_array = {}
for _dir in dir_list:
    pred_array[_dir] = {}
    for _seed in seed_list:
        pred_array[_dir][_seed] = {}
        for _csv in csv_list:
            pred_array[_dir][_seed][_csv] = {}

            # seed123だけを学習し、他はテストとする
            df = csv_array[_dir][_seed][_csv].copy()
            y = df['people']
            x = df[['time', 'area']]
            
            if _seed == '123':
                model = get_model(x, y)

            else:
                df['pred'] = model.predict(x)
                pred_array[_dir][_seed][_csv] = df
```
