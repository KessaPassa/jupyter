# 変化率 = 各メッシュの人数 / 全体の人数

```python
def get_ratio_per_sum(base):
    row_sum = base.sum()
    for index in range(len(base.columns)):
        base.iloc[:, index] /= base.iloc[:, index].sum()
        
    base = base.append(row_sum, ignore_index=True)
    base.rename(index={len(base.index)-1: 'sum'}, inplace=True)
    base = base.round(4)
    base = base.applymap(lambda x: np.nan if x <= 0.001 else x)
    base.dropna(how='all', inplace=True)
    
    return base
```

```python
def get_ratio_per_time(base):
    copy = base.copy()
    for _times in range(6 - 1):
        base.iloc[:, _times+1] = copy.iloc[:, _times+1] / copy.iloc[:, _times]
#         base.iloc[:, 0] = base.iloc[:, 0].apply(lambda x: 0 if x == 0 else 1)
    base = base.round(4)
    
    return base
```

```python
df_mobile = {}
df_census = {}
for _dir in dir_list:
    df_mobile[_dir] = {}
    df_census[_dir] = {}
    for _seed in seed_list:
        df = csv_array[_dir][_seed]['mobile'].copy()
        df_mobile[_dir][_seed] = get_ratio_per_sum(df)

        df = csv_array[_dir][_seed]['census'].copy()
        df_census[_dir][_seed] = get_ratio_per_sum(df)
```
